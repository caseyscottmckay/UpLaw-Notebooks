```python
from fastai.tabular import *
import pandas as pd
import numpy as np
#from google.colab import drive
#drive.mount('/content/drive')
#path = Path('/content/drive/My Drive/data/scdb')
path = Path('/home/jupyter/data')
df = pd.read_csv(path/'scdb/SCDB_2019_01_caseCentered_Citation.csv',encoding='iso-8859-1').fillna(0)
df.head()

dep_var='caseDisposition'
cat_names=['certReason','lcDisposition','lcDispositionDirection','caseOrigin','caseOriginState','caseSource','jurisdiction','issue','issueArea','lawType','majOpinAssigner','majOpinWriter', 'petitioner', 'respondent']
cont_names=['term', 'minVotes', 'majVotes']
procs = [FillMissing, Categorify, Normalize]
test = TabularList.from_df(df.iloc[7000:8000].copy(), path = path, cat_names=cat_names, cont_names=cont_names)
data = (TabularList.from_df(df, path=path, cat_names=cat_names, cont_names=cont_names, procs=procs)
  .split_by_idx(list(range(7000,8000)))
  .label_from_df(cols=dep_var)
  .add_test(test)
  .databunch())

data.show_batch(rows=10)
learn = tabular_learner(data, layers=[200,100], metrics=accuracy)

#unfreezing, fine-tuning, and learning rates

learn.fit_one_cycle(2)
learn.save('stage-1')
learn.unfreeze()
learn.fit_one_cycle(1)
learn.load('stage-1') 
learn.lr_find()
learn.recorder.plot()
learn.unfreeze()#train whole model
learn.fit_one_cycle(1, 1e-2)
#learn.fit_one_cycle(2, max_lr=slice(1e-4, 1e-3)) #2nd part is 10x less than bad, first part is well before things went bad, 
learn.export(path/'scdb_tabular_prediction_export.pkl')
learn.predict(df.iloc(1))
```
